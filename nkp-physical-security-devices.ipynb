######  2. Motion Detection and Video Surveillance
***

After completing this topic, you should be able to: ``recognize when to implement motion detection and video surveillance``

Now, in this presentation, we'll take a look at several of the components that you might find in a security control center or a monitoring station. And with respect to what is being monitored in this regard, we are more so talking about physical activity.

So not so much accessing a file or a folder in an unauthorized fashion or hacking in through a firewall. So rather that physical access, so are people quite literally to get into any given environment and/or out without being detected. 

So generally, you do want to keep an eye on that sort of thing. And you can expect to find something like a CCTV, or closed circuit TV system, which provides active monitoring and surveillance, usually 24/7.

Now, there might also be an intrusion detection system of some kind, but again, not really a network intrusion detection system. It's not monitoring network packets. It is quite literally monitoring if people are able to break into an environment. And of course, there would also be alarm monitoring associated with that, if of course they are able to break in. 

So the components with respect to closed circuit TV, of course, involves the camera itself. There may be various other recorders, including audio recorders, but there has to be something that records the actual activity. And the monitoring station is typically where you would see all of the actual screens, the monitors for each camera. And you might see them switching from one view to another, things along those lines. 

But with respect to the usage and/or the modes, clearly this is implemented for surveillance, but might also be for assessment. In some cases there is a lot of activity that is expected and normal, of course. But you might be trying to assess something that appears to be abnormal. 

It's a fairly effective deterrent, of course. And it might also be used for evidentiary archives. Just keeping that data so you can go back and view on any given date to see what happened, if it is discovered that something was compromised.

######  Cameras
***

Now, with respect to the cameras themselves, there are a few different types. Of course an outdoor camera is usually mounted to the side of a building or something along those lines, generally has a fixed position. Now, there can be a few different types, but again, it simply means that the camera is not mobile in any way, shape, or form.

A PTZ camera is short for pan, tilt and zoom. So even though it might be fixed, it can still rotate on a pivot, for example. And it can zoom in to get better resolution and things like that. 

You might see dome cameras, little more on the inside of most environments. You might see them, for example, on the ceiling of a bank and any other type of secure environment. And really any kind of IP camera simply means that it is attached to a network. So the entire network of cameras can communicate and send their images to a central location. 

< [Table of Contents](#top) | [References](#references) >

***
###### Cameras - Terms
***

Some key terms with respect to the camera itself. You can expect to see a value such as FPS, or frames per second. And the more frames per second you can capture, the greater the detail and the more data you can expect to gather. But you generally want to see a value that's a happy medium. Because, again, if you are capturing too many frames, that's a lot of data. And if you're just trying to capture a standard physical intrusion, you generally don't need very high values of frames per second. Standard video, a regular video camera, I believe is around 30 frames per second. So you can usually get by with values lower than that. 

Certainly the resolution if you need to capture facial images, for example. You would want very high resolution, particularly if it's far away. 

The lighting level certainly needs to be considered. Because if you anticipate people trying to intrude at night, for example, then you clearly want the area that is being monitored to be well lit. 

Compression, because clearly you are storing a lot of data. So you would want to see that compressed. And then typically there's some sort of digital video recorder, or DVR, that actually records what is happening.

###### Cameras - Motion Detection
***

Now when it comes to motion detection, the capability of a surveillance system capture any given event at the point motion is detected by that system, is clearly what we are referring to. But most of them use a software based monitoring algorithm to detect the motion. And there are certain degrees, if you will, with respect to what will trigger any given motion. And you might be familiar with something like a home system that is motion detecting, but has a threshold, for example, that you can configure. Because if you have pets, for example, the assumption is that they are going to move around throughout the house while you are gone. And if it was set off by any kind of motion, clearly you'd be getting alarms all the time. So most of them are configurable with respect to what is the threshold of maybe the size of the object, or how much it moves. 

But ultimately, it is really the idea of starting the capture process only when motion is detected. Because we don't really need to capture anything when there is nothing going on. Now this might also be referred to as activity detection, because, of course, motion is an activity. 

But coming back to the more advanced motion detection surveillance systems that I referred to earlier, how some of them can be configured with thresholds. You might have this ability to implement a configuration that says, well, it's going to actually try to determine whether or not to initiate an alarm, based on the activity. 

And again, I mentioned the home system that can be configured for sort of a size threshold, if we're talking about pets. Now the pets aren't very large. In most cases it's a dog or a cat. 

So you might be able to configure some kind of threshold with respect to the size of the object that is being monitored. But if it's a person, well, that's significantly larger. So it can make that determination, as to whether or not this is actually suspicious movement. 

Now there is also video content and analysis and even motion path analysis that can also be implemented. And this is detection based on the pixel change and/or video content analysis or even motion path analysis. And again, the idea here is that you typically want the system to be up and running 24/7 all the time. But if it alarmed every single time it captured any kind of motion, it would just keep going all the time in a lot of environments. So again, you can configure certain values to say that this is normal. Whenever it, quote, sees this kind of activity, then that can be considered normal. 

Perhaps in a bank environment, if somebody always goes into the vault using the sort of the same path and the same general motions, and maybe even at the same time of day, those things can be considered. And you can, in essence, say that this is normal, this is acceptable. Do not fire off an alarm in that event. But that's up to you. How you configure the system is certainly your call. But it is generally something to be considered. But in most cases, you would probably error on the side of caution. If it's not sure, then let's alarm just to be safe.

######  3. Asset Tracking
***

After completing this topic, you should be able to: ``identify when to implement asset tracking``

In this presentation, we'll take a look at managing and maintaining asset inventory records. 

And with respect [Asset Inventory Records. ] to what is generally recorded as part of any given organization's assets, it's usually more so hardware and software. As they're considered to be a little more crucial to being able to successfully implement security and risk management strategies. 

But there may be other things that are included as well, but that is certainly going to vary depending on the organization. 

So when it comes to marking or indicating any given asset, it should include the following, some sort of unique asset identifier which is typically a tag of some kind. So that of course it is uniquely identified amongst every asset. You should never have duplicate identifiers in any situation. 

Then it should also include some sort of qualitative and/or quantitative value, and that of course can include monetary value. How critical it is to the environment. And/or whether or not it is considered to be a high value asset. 

It should also include components such as the location, where it is geographically, whether it's part of call it a headquarters or a remote site or a branch office. 

And again, the security classification or the sensitivity level. That kind of relates to its value, of course. But it certainly is also something that should be recorded. In essence, how severe is it if this particular asset was lost or compromised. 

It should also include an asset group if applicable. There might be certain assets of course that are part of a category, for example, all desktop computers, all laptop computers, all mobile devices, things along those lines. 

It should indicate who is the owner and/or the custodian of the asset and the recovery priority. So if it is damaged, lost or stolen, you know, how quickly it should be replaced, what kind of severity level does its loss represent, things along those lines.

######  Software Asset Management
***

[Software Asset Management. ] And then when it comes to software asset management, this is of course used to identify and track an organization's operating systems. 

The software applications that are in use. 

The mobile device operating systems and the apps that are installed. 

Any cloud applications that are being used, and possibly any virtualization that is being implemented. 

And with respect to the inventory, keeping this up to date can be manual or automated, it depends on the system and your approach. But it should be cross-platform, because if you do have multiple different operating systems and different vendors. It should be able to operate and/or be accessible from just about any operating system, so something that is maybe web-based might be ideal. But, again, this is going to vary. 

And licensing should also be considered, and you typically have an inventory of licenses as well. So when you purchase software for example, it defines that you can install it x number of times. So you can maintain an inventory of how many of those installations have occurred and or how many licences you have left. 

It should also include audit information with respect to how often this is reviewed. 

And perhaps maybe the history of the application, the versions that have been implemented. 

Any overages with respect to tracking that usage. Again, you may have exceeded the installations in some cases. Something that indicates if there was any piracy that took place with respect to the software. 

And if any virtualization is being implemented, in some cases, there are unique licensing conditions that apply to virtualization. Some examples include the fact that if you are running it virtually then there might only be a single copy running on a system that is accessed, perhaps using something like remote desktop. And as such, you only actually install it once, but it is still being used by multiple different users. So situations like that might have unique conditions in the licensing agreement. 

And of course, the mobile devices as well. They certainly have software on them as well, so that should also be part of the overall process.

######  Social Engineering
***

Now in this presentation, we'll take a look at several different implementations of security devices and or general approaches to security. We're going to begin with what's called social engineering. The social engineering is simply when a threat actor is able to get around whatever security barriers you put in place by getting flesh and blood people to do the work for them. If you've ever watched a heist movie, you know that a con-man is need to convince people to let down their guard. Unfortunately, there's a thousand ways to do this. So it's not like blocking a port on a Firewall. We have to educate our users on how to recognize valid people who are allowed to work on the network equipment. How to respond to phone calls, emails and texts that might seem to becoming from someone with authority. And what to do if they find a USB drive just lying around. We want to mention this first because every form of security can be undone [Social Engineering includes attacks based on human interaction that can circumvent nearly all other protections.
